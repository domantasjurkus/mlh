```python
import numpy as np
import pylab as plt
%matplotlib inline

# Datum
trainX = np.loadtxt('X_train.csv', delimiter=',', skiprows=1)
traint = np.loadtxt('y_train.csv', delimiter=',', skiprows=1)
traint = traint[:,1][:,None]
#print("Train X shape:", trainX.shape)
#print("Train t shape:", traint.shape)

# Need to split data into validation set
#split_index = int(trainX.shape[0]/cross_validation_folds)
#print(split_index)
cvX = None
cvt = None

testdata = np.loadtxt('X_test.csv', delimiter=',', skiprows=1)
testX = testdata[:,:2]
testt = testdata[:,2]

def cv(trainX, traint, folds=6, K=3):
    slice_size = int(trainX.shape[0]/folds)
    total_accuracy = 0
    
    for i in range(folds):
        print("CV fold:", i)
        split_start = slice_size * i
        split_end = slice_size * (i+1)
        training_X = np.concatenate((trainX[:split_start], trainX[split_end:]))
        training_t = np.concatenate((traint[:split_start], traint[split_end:]))
        testing_X = trainX[split_start:split_end]
        testing_t = traint[split_start:split_end]
        
        hits = 0
        for i in range(slice_size):
            c = knn_classifier(trainX, traint, testing_X[i,:], K)
            if (c == testing_t[i]):
                hits += 1
        accuracy = hits/testing_t.shape[0]
        total_accuracy += accuracy
    
    total_accuracy /= folds
    return total_accuracy
    

def knn_classifier(trainX, traint, testrow, K=3):
    distances = ((trainX - testrow)**2).sum(axis=1)
    dc = zip(distances,traint)
    dc = sorted(dc,key = lambda x:x[0])
    classes = []
    votes = []
    for k in range(K):
        this_class = dc[k][1] # get the class of the kth ranked
        if not this_class in classes:
            classes.append(this_class)
            votes.append(1)
        else:
            index = classes.index(this_class)
            votes[index] += 1
    best_class = classes[0]
    best_vote = votes[0]
    pos = 1
    for cl in classes[1:]:
        if votes[pos] > best_vote:
            best_vote = votes[pos]
            best_class = cl
        pos += 1
    return best_class

Ks = np.arange(1,10,1)
for k in Ks:
    correct = 0
    for i,row in enumerate(testX):
        c = knn_classifier(trainX,traint,row,K=k)
        if c == testt[i]:
            correct += 1
    accuracy.append(1.0*correct / (1.0*len(testX)))
cv(trainX, traint)

#test_index = 150
#c = knn_classifier(trainX, traint, testX[test_index,:], K=8)
```
